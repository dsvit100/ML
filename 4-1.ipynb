```python
import pandas as pd
import numpy as np

df = pd.read_csv('data/Fish.csv')
df.head()
```

```python
df['Species'].unique()
# 유니크 함수는 중복되지 않은 컬럼의 값들을 출력해줌
```

```python
fish_input = df[['Weight', 'Length2', 'Length3', 'Height', 'Width']]
fish_input.head()
```

```python
fish_target = df[['Species']]
```

```python
# 트레이닝, 테스트로 데이터 분리
from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target = \
train_test_split(fish_input, fish_target)
```

```python
# 각 데이터들간의 거리 수치가 다르기 때문에 스케일링 해줘야 함

from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
ss.fit(train_input)
train_scaled = ss.transform(train_input)
test_scaled = ss.transform(test_input)
```

```python
type(test_scaled)
```

```python
from sklearn.neighbors import KNeighborsClassifier

kn = KNeighborsClassifier()
kn.fit(train_scaled, train_target)
```

```python
print(kn.score(train_scaled, train_target))
print(kn.score(test_scaled, test_target))
```

```python
test_scaled[:5]
```

```python
kn.predict(test_scaled)
```

```python
# 위의 예상치가 어떤 근거로 출력이 된 것인지 알 수 있는 함수
# 'Bream', 'Roach', 'Whitefish', 'Parkki', 'Perch', 'Pike', 'Smelt'

kn.predict_proba(test_scaled)

# 7개의 클래스 중 하나일 확률
# 1열 기준 Whitefish일 확률 80%, Perch일 확률 20%
```

```python
distances, indexes = kn.kneighbors(test_scaled[3:4])
print(distances)
print(indexes)
# print(test_target)
# 판다스는 인덱스를 컬럼으로 인식, 넘파이는 인덱스를 정수로 인식하기 때문에

print(train_target.iloc[indexes[0]])
# 2차원 배열을 1차원으로 만들기 위해 iloc 사용
# test_scaled 4번째 값의 최근접을 뽑기 위한 것이기 때문에
# 근거는 train에 있음
```

## 로지스틱 회귀
- 결과값은 숫자로 나옴 = 회귀
- z 값이 높아지면 그 데이터일 확률이 높아진다 의 의미로 분류모델
- 시그모이드 함수 = 수식으로부터 나온 그래프라고 생각하면 됨
- 시그도모이드 함수에서 100%란 존재할 수 없다.
- z값이 많아져도 0이나 1이 될 수는 없다. 어떤 데이터를 쓰더라도 0과 1에 수렴한다.

### 5개의 특성을 통해 bream인지 smelt인지

```python
bream_smelt_indexes = (train_target == 'Bream') | (train_target == 'Smelt') #~거나 ~거나
# bream이거나 smelt인 데이터를 모두 True로 바꿔줌

bream_smelt_indexes.head()
```

```python
train_bs = train_scaled[bream_smelt_indexes['Species']]
# # train_scaled에서 bream과 smelt 값(무게,길이 등)만 뽑아옴

# train_bs
train_scaled
```

```python
target_bs = train_target[bream_smelt_indexes['Species']]

# target_bs
train_target
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(train_bs, target_bs)
```

```python
print(lr.predict(train_bs[:5]))
print(lr.predict_proba(train_bs[:5]).round(3))
# train_bs까지가 array, .round를 통해 소숫점 뒤 세자리까지 출력
```

```python
lr.coef_ #코이피션트
# z = -0.5 x (Weight) + -0.6 x (Length) + ... 의 회귀 방정식을 만들어냄
# 생선을 판별하는 z 값을 구하는 방정식
```

```python
lr = LogisticRegression(C=20, max_iter=1000)
# LogisticRegression() 안 넣어도 학습은 가능
# 사람이 직접 입력해야 하는 파라미터가 있음
# max_iter은 몇 번을 반복할까요? (기본값 = 100)
# C는 숫자가 커질수록 규제가 약해짐
lr.fit(train_scaled, train_target) # 7마리 생선이 다 들어가 있음

print(lr.score(train_scaled, train_target))
print(lr.score(test_scaled, test_target))
```

```python
lr.predict_proba(test_scaled[:5]).round(3)
# 'Bream', 'Roach', 'Whitefish', 'Parkki', 'Perch', 'Pike', 'Smelt' 이 순서
```

```python
# softmax 적용 전
decision = lr.decision_function(test_scaled[:5])
print(np.round(decision, decimals=2))
```

```python
lr.coef_.shape
# coef 특성을 만들어 낼 때 사용
# 각 분류종에 대한 방정식을 각각 만들어냄 = 7개

# 로지스틱 함수는 각 z의 
```

- 로지스틱회귀 = z값을 만들어주는 것
- 2진분류에서는 = 하나만 계산하고 나머지는 -하나
- 다중분류에서 로지스틱 회귀는 사용 X
- 왜? z1... zn개의 값이 나오고 각 z값의 확률을 계산하기 때문에 n개 클래스의 전체 확률을 계산했을 때, 합이 100%가 넘어감 (비율이 적용되지 않음)
- 따라서 이 경우 소프트맥스 = 데이터가 여러개일 때
