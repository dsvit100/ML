```python
import pandas as pd

wine = pd.read_csv('https://bit.ly/wine_csv_data')
wine.head()
```

```python
data = wine[['alcohol', 'sugar', 'pH']]
target = wine[['class']]
```

```python
from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target = \
train_test_split(data, target)
```

```python
# sub_input, val_input, sub_target, val_target = \
train2_input, test2_input, train2_target, test2_target = \
train_test_split(train_input, train_target)
```

```python
train2_input.shape
# 학습용 데이터가 약 3600개
```

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier()
dt.fit(train2_input, train2_target)
print(dt.score(train2_input, train2_target))
print(dt.score(test2_input, test2_target))
```

```python
print(dt.score(test_input, test_target))
```

### 교차검증(k-폴드 교차검증)
- 전체 데이터의 1/k를 k번만큼 테스트 데이터로 사용하여 교차적으로 모델을 검증
- 보통 기본인 5-폴드 교차검증이나 10-폴드 교차검증을 사용

```python
from sklearn.model_selection import cross_validate
# 기본적으로 1/5씩 교차검증을 수행함 (총 5회)

scores = cross_validate(dt, train_input, train_target)
# cross_validate(평가할 모델 객체, 훈련 세트 전체(input, target))
print(scores)
```

## 하이퍼파라미터 튜닝
- 매개변수를 자동화로 할 수는 없을까?

```python
from sklearn.model_selection import GridSearchCV

# GridSearchCV 에서는 파라미터를 딕셔너리 값으로 해야함
params = {
    'min_impurity_decrease': [0.0001, 0.0002, 0.0003, 0.0004]
    # 최소불순도
}
dt = DecisionTreeClassifier()
```

```python
gs = GridSearchCV(dt, params)

# GridSearchCV = 교차검증과 동시에 적절한 하이퍼파라미터 값을 골라주는 클래스
# GridSearchCV(사용객체, 파라미터)
# n_jobs 안 써도 동작함
# 기본적으로 n_jobs=1, 하나의 cpu만 사용하는 것으로 되어있음
# n_job=-1로 설정하면 최대한도 CPU를 사용하여 속도를 높일 수 있음
```

```python
# train_target.info()
gs.fit(train_input, train_target)
```

```python
gs.best_estimator_
# 최적의 조합
```
